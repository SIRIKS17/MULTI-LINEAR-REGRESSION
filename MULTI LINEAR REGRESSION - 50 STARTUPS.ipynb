# Model building

```python
import statsmodels.formula.api as smf 
model = smf.ols('Profit~RDS+ADMS+MKTS',data=data1).fit()
model
```

```python
model.params
```

```python
model.tvalues , np.round(model.pvalues,5)
```

```python
(model.rsquared,model.rsquared_adj)
```

# Regression model 

```python
slr_a=smf.ols('Profit~ADMS',data=data1).fit()  
print(slr_a.tvalues, '\n', slr_a.pvalues)
```

```python
slr_m=smf.ols('Profit~MKTS',data=data1).fit()  
print(slr_m.tvalues, '\n', slr_m.pvalues)
```

```python
mlr_am=smf.ols('Profit~MKTS+ADMS',data=data1).fit()  
print(mlr_am.tvalues, '\n', mlr_am.pvalues)  
```

# Collinearity 

```python
rsq_r=smf.ols("RDS~ADMS+MKTS",data=data1).fit().rsquared
vif_r=1/(1-rsq_r)
rsq_a=smf.ols("ADMS~RDS+MKTS",data=data1).fit().rsquared
vif_a=1/(1-rsq_a)
rsq_m=smf.ols("MKTS~RDS+ADMS",data=data1).fit().rsquared
vif_m=1/(1-rsq_m)
d1={'Variables':['RDS','ADMS','MKTS'],'Vif':[vif_r,vif_a,vif_m]}
Vif_df=pd.DataFrame(d1)
Vif_df
```

# Residual for normality

```python
import statsmodels.api as sm
qqplot=sm.qqplot(model.resid,line='q') # line = 45 to draw the diagnoal line
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
list(np.where(model.resid<-30000))
```

# Residual for homoscedasticity

```python
def get_standardized_values( vals ):
    return (vals - vals.mean())/vals.std()   
```

```python
plt.scatter(get_standardized_values(model.fittedvalues),
            get_standardized_values(model.resid))
plt.title('Residual Plot')
plt.xlabel('Standardized Fitted values')
plt.ylabel('Standardized residual values')
plt.show()
```

# Cook's distance
# model deletion

```python
model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance
c
```

```python
fig = plt.subplots(figsize=(15,10))
plt.stem(np.arange(len(data1)), np.round(c, 3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
(np.argmax(c),np.max(c))
```
